```python
import pandas as pd
from pathlib import Path
import zipfile

# 根目录
DATA_DIR = Path("../data/datasets_full.csv")

# 指向 genuine 用户压缩包
zip_path = DATA_DIR / "genuine_accounts.csv.zip"

# 打开 zip
with zipfile.ZipFile(zip_path) as z:
    # 列出 zip 内文件
    print(z.namelist())
    # 假设里面有 'genuine_accounts.csv/tweets.csv'
    file_name = 'genuine_accounts.csv/tweets.csv'
    
    # 读取 CSV
    df_genuine_tweets = pd.read_csv(z.open(file_name), encoding="latin1")
    
df_genuine_tweets.head()
```

检查一下数据都有些啥特征

```python
import pandas as pd
from pathlib import Path
import zipfile

DATA_DIR = Path("../data/datasets_full.csv")
zip_path_genuine = DATA_DIR / "genuine_accounts.csv.zip"

# users.csv
with zipfile.ZipFile(zip_path_genuine) as z:
    users_genuine = pd.read_csv(z.open("genuine_accounts.csv/users.csv"), encoding="latin1")
    
# tweets.csv
with zipfile.ZipFile(zip_path_genuine) as z:
    tweets_genuine = pd.read_csv(z.open("genuine_accounts.csv/tweets.csv"), encoding="latin1", low_memory=False)

# 用户特征
print("Users columns:")
print(users_genuine.columns)
# print(users_genuine.head())

# 推文特征
print("\nTweets columns:")
print(tweets_genuine.columns)
# print(tweets_genuine.head())
```
